```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv( 'D:\\КУРСЫ DATA SCIENCE\\МОИ ЗАДАЧКИ\\iris.csv')
df
```

```python
df = df \
    .rename (columns = 
            {'sepal length' : 'sepal_length',
              'sepal width' : 'sepal_width',
              'petal length' : 'petal_length',
              'petal width' : 'petal_width'})
df
```

```python

sns.distplot(df['petal_width'], color = "blue")
sns.distplot(df['petal_length'], color ="green")
sns.distplot(df['sepal_width'], color = "yellow")
sns.distplot(df['sepal_length'], color = "orange")
```

```python
fig, axes = plt.subplots(1, 1, figsize=(10,10))
sns.distplot( df["sepal_length"] , color="skyblue", label="Sepal Length", ax=axes)
```

```python
axes.legend()
```

```python
sns.distplot( df["sepal_width"] , color="green", label="Sepal Width", ax=axes)
```
