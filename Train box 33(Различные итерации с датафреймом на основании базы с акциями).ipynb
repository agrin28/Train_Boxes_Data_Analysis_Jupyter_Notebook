Задача с ценами акций на бирже

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
stock=pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/amzn_stock.csv', index_col='Date', parse_dates=True)
stock.head( )
#сделали так, чтобы каждое значение даты в фрейме стало индексом , только с тайм сириес данными
```

```python
stock.index
```

```python
stock['2010']
# теперь можно индексироваться по указанной дате ( например, отобрали все за 2010 год)
```

```python
stock['2010-02':'2011-03']
# либо отобрали период с февраля 2010 - по март 2011 года
```

```python
stock.resample('2h').asfreq()
# resample('2h') позволяет разбить дату на другие промежутки времени, например, в датасете цены по дням распределены, а можно сделать по часам
# НО значений по часам нет, т.к. не было в исходном датафрейме этих данных
```

```python
stock.resample('1w').mean()
# данные за неделю (среднее значение)
```

```python
stock.head(10)
```

```python
stock.rolling(3)
# Сместились на 3 строчки вниз (катающееся окно)
```

```python
stock.rolling(3).mean()
# среднее значение строк, опираясь только на значения первых 3х строк. 4 и 5 дней без значений, 
# потому что не хватает данных перед ними , 3его, 2ого дня для расчета
```

```python
stock.rolling(3, min_periods=1).mean()
# min_periods=1 делает так, что расчет среднего идет по предидущего значению, т.е. для среднего значения хватает 1 числа
```

```python
stock.expanding(3).mean()
# среднее значение по предидущим значениям, например для первой колонки остается, для 2ой первая, для 3ей - 1 и 2ая, для 4 - 1,2 и 3 и т.д.
```

```python
stock.ewm(alpha=0.7).mean( )
```

```python
stock['Open'].plot()
```

```python
ns=stock['Open'].rolling(10, min_periods=1).mean()
# сглаживание функции
stock['Open'].plot()
ns.plot()
```
